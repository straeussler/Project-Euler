```python
#ugly solution
```

```python
p = 3
max_len = 0
while p < 1000:
    list = []
    for a in range(1,998):
        for b in range(1,998):
            if (a**2 + b**2)**0.5 == p -a -b:
                if a not in list:
                    list.append(a)
                    list.append(b)
                    list.append(p-a-b)
                    print(list)
            if len(list) > max_len:
                max_len = len(list)
    p += 1
print(max_len//3)

```
