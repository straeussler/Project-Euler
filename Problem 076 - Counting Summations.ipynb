```python
nums = []
for x in range(1,101):
    nums.append(x)
target = 100
matrix = {}

for y in range(0, target+1):
    matrix[y,0] = 1
    
for y in range(0, target+1):
    print(y, ":", 1, end = ' ')
    for x in range(1, len(nums)):
        matrix[y,x] = 0
        if y >= nums[x]:
            matrix[y,x] += matrix[y,x-1]
            matrix[y,x] += matrix[y-nums[x],x]
        else:
            matrix[y,x] = matrix[y,x-1]
        print(matrix[y,x], end = ' ')
    print()
print('solution:',matrix[target,len(nums)-2])
```
